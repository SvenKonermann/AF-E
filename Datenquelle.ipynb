```python
# Import der benötigten Libraries 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import os

pd.set_option('display.max_colwidth', 50)

# Working-Directory
os.chdir('C:/Users/Sven Konermann/Documents/Master/2_Semester/aF_E_Faelle/Unterlagen/00_all')
```

```python
# Files einlesen (.csv)
# Dataset 1
data1 = pd.read_csv("./Subset_IT_Projektmanagement.csv", 
                    sep=";", 
                    encoding="utf-8",
                    low_memory=False)
# Dataset 2
data2 = pd.read_csv("./Subset_IT_Consulting.csv", 
                    sep=";", 
                    encoding="utf-8",
                    low_memory=False)
# Dataset 3
data3 = pd.read_csv("./Subset_HR.csv", 
                    sep=";", 
                   encoding="utf-8",
                    low_memory=False)
# Dataset 4
data4 = pd.read_csv("./Subset_Data_Science.csv", 
                    sep=";", 
                    encoding="utf-8",
                    low_memory=False)
# Dataset 5
data5 = pd.read_csv("./Subset_Business_Analyst.csv", 
                    sep=";", 
                    encoding="utf-8",
                    low_memory=False)
# Dataset 6
data6 = pd.read_csv("./Subset_ApplManager.csv", 
                    sep=";", 
                    encoding="utf-8",
                    low_memory=False)
# Dataset 7
data7 = pd.read_csv("./Subset_Accounting.csv", 
                    sep=";", 
                   encoding="utf-8",
                    low_memory=False)
```

```python
# Anzahl Datensätze: 5'583'709
number_of_rows_data1 = len(data1)
number_of_rows_data2 = len(data2)
number_of_rows_data3 = len(data3)
number_of_rows_data4 = len(data4)
number_of_rows_data5 = len(data5)
number_of_rows_data6 = len(data6)
number_of_rows_data7 = len(data7)
print(number_of_rows_data1)
print(number_of_rows_data2)
print(number_of_rows_data3)
print(number_of_rows_data4)
print(number_of_rows_data5)
print(number_of_rows_data6)
print(number_of_rows_data7)
```

```python
# Zusammenfügen der Objekte
frames = [data1, data2, data3, data4, data5, data6, data7]
result = pd.concat(frames)
number_of_rows = len(result)
print(number_of_rows)
```

```python
# Datum als Datetime abspeichern
result['active_start_on'] = result['active_start_on'].astype('datetime64')
result['active_end_on']   = result['active_end_on'].astype('datetime64')
result['duration']        = result['duration'].astype('int')

# Spalte mit Jahr erstellen
result['year'] = result['active_start_on'].dt.year

result.head(5)
```

```python
# Visualisierung der Jobkategorien mit der Anzahl an Inseraten pro Kategorie
result['lvl1_jobcategory_name'].value_counts()[:19].plot(kind='barh')
plt.savefig('my_fig.pdf', dpi=100, bbox_inches="tight")
```

```python
# Visualisierung Jobkategorien in Prozent vom Total
data = result['lvl1_jobcategory_name'].value_counts()
data = data / data.sum() * 100

fig, ax = plt.subplots()
y = data.index
x = data.values

ax.barh(y, x, align='center')
ax.set_yticks(y)
ax.set_xlabel('Prozent')

# plt.savefig('my_fig.pdf', dpi=100, bbox_inches="tight")
plt.show()

result.to_csv("Subset_all.csv", sep=";")
```
