```python
from bs4 import BeautifulSoup
```

```python
url = 'https://sonnenaufgang.online/sun/uster'
```

```python
response = requests.get(url)
```

```python
print(response)
```

```python
print(response.text)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
sunrise_data = soup.find('span', {'data-name': 'sunrise'})
```

```python
print(sunrise_data)
```

```python
sunrise_data = sunrise_data.text
```

```python
print(sunrise_data)
```

```python
sunrise_data = sunrise_data[:-3]
```

```python
print(sunrise_data)
```

```python
print(f'Die Sonne geht heute um {sunrise_data} Uhr auf.')
```
